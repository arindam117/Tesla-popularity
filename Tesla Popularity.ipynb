```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

# This will help fetch data from google trends
from pytrends.request import TrendReq   

# Creating a pytrend object to request data from google trends
# hl is host language
pytrends = TrendReq(hl='en-us')  


# Extract data about keywords 
keywords = ['Tesla']

# The timeframe = 'today 5-y' gives data back to 5 years from today
pytrends.build_payload(keywords,timeframe = 'today 5-y') 

# This uses the .interest_over_time() module to give a cummulative estimate 

data = pytrends.interest_over_time()


data.plot(kind='line')   

# This gives the heading / title of the plot
plt.title('Trends in Tesla searches')  

# This renames the X-axis label
plt.xlabel('Year Stamps') 

# This renames the Y=axis lables
plt.ylabel('Average Enquiries') 

data2 = pytrends.interest_by_region(resolution='COUNTRY',inc_low_vol=True)

# THIS GIES THE COUNTRIES WITH THE LARGEST SEARCH ENTRIES ABOUT TESLA 
data2 = data2['Tesla'].nlargest(10)

# DATA FRAME DIVIDED BY COUNTRIES

data2 = data2.head(20)

data2 = data2.to_frame()

# PLOTTING THE DATAFRAME USING PANDAS IN BAR FORMAT 

data2.plot(kind='bar')

# using matplotlib we can do further modification 

plt.title('Top countries which searched Tesla')
plt.xlabel('Country Names')
plt.ylabel('Search Counts')

plt.show()
```
